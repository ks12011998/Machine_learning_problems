```python
img_dir = '/kaggle/input/age-classification/Age_detection/train_DETg9GD/Train/'
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tqdm import tqdm
```

```python
train_image = []
for i in tqdm(range(train.shape[0])):
    img = image.load_img(img_dir+train['ID'][i], target_size=(28,28,3), grayscale=False)
    img = image.img_to_array(img)
    img = img/255
    train_image.append(img)
X = np.array(train_image)
```

```python
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
train['Class']= label_encoder.fit_transform(train['Class']) 
```

```python
y = train['Class']
y
```

```python
y = to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
#importing the keras libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense

#Initailising the CNN
classifier = Sequential()
#Step1-Convolution
classifier.add(Convolution2D(32,3,3,input_shape=(28,28,3),activation = 'relu'))

#Step2 - Pooling
classifier.add(MaxPooling2D(pool_size=(2,2)))

#Flattening
classifier.add(Flatten())

#Full connection
classifier.add(Dense(128,activation = 'relu'))
classifier.add(Dense(3,activation = 'softmax'))

#Compiling the CNN
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy' , metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, epochs=20,steps_per_epoch=2654, validation_data=(X_test, y_test))
```

```python
prediction = classifier.predict_classes(test,verbose=2)
```

```python
final_list = []
for i in range(len(prediction)):
    if prediction[i] == 0:
        final_list.append('MIDDLE')
    elif prediction[i] == 1:
        final_list.append('YOUNG')
    else:
        final_list.append('OLD')
```
